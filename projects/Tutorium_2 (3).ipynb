```python
#Diskrete Gleichverteilung
x=np.random.randint(0,10,size=10000)
sns.distplot(x,kde=False, rug=False)
```

```python
#Binomialverteilung
x=np.random.binomial(n=50,p=0.2,size=10000)
sns.distplot(x,kde=False, rug=False)
```

```python
#Poissonverteilung
x=np.random.poisson(lam=3,size=10000)
sns.distplot(x,kde=False, rug=False)
```

```python
#Hypergeometrische Verteilung
x=np.random.hypergeometric(ngood=30,nbad=200,nsample=20,size=10000)
sns.distplot(x,kde=False, rug=False)
```

```python
#Stetige Gleichverteilung
x=np.random.uniform(low=0,high=10,size=10000)
sns.distplot(x)
```

```python
#Normalverteilung
x1=np.random.normal(loc=20,scale=3,size=10000)
x2=np.random.normal(loc=20,scale=10,size=10000)
sns.distplot(x1, hist=False, rug=False)
sns.distplot(x2, hist=False, rug=False)
```

```python
#Standardnormalverteilung
x=np.random.normal(size=10000)
sns.distplot(x)
```

```python
#t-Verteilung
x=np.random.standard_t(df=7,size=10000)
sns.distplot(x)
```

```python
#Chi2-Verteilung
x=np.random.chisquare(df=3,size=10000)
sns.distplot(x)
```
